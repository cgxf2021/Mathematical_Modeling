```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fun1 = lambda x: -0.04658 * x ** 3 + 0.5494 * x ** 2 - 1.3151 * x + 1.415
fun2 = lambda x: 0.1545 * x ** 3 - 0.799 * x ** 2 + 0.4316 * x + 3.744

A = np.random.uniform(low = 0.5, high = 3.2, size = (1, 70))
B = np.zeros(shape = (1, 70))
for i in range(70):
    B[0, i] = np.random.uniform(low = fun1(A[0, i]), high = fun2(A[0, i]))
```

```python
# 生成画布
fig = plt.figure(figsize = (8, 7), dpi = 100)

# 添加绘图区域
ax = fig.add_subplot(111)

# 设置标题
ax.set_title("Anacomy of a figure", fontsize = 20)

# 设置刻度
ax.set_xlim(0, 4)                   # 设置x轴范围
ax.set_ylim(0, 4)                   # 设置y轴范围

xMajorLocator = plt.MultipleLocator(1)              # x主刻度线间隔1
ax.xaxis.set_major_locator(xMajorLocator)
yMajorLocator = plt.MultipleLocator(1)              # y主刻度线间隔1
ax.yaxis.set_major_locator(yMajorLocator)

xMinorLocator = plt.MultipleLocator(0.25)           # x副刻度线间隔0.25
ax.xaxis.set_minor_locator(xMinorLocator)
yMinorLocator = plt.MultipleLocator(0.25)           # y副刻度线间隔0.25
ax.yaxis.set_minor_locator(yMinorLocator)

xMinorFormatter = plt.FormatStrFormatter('%0.2f')   # 显示x副刻度线
ax.xaxis.set_minor_formatter(xMinorFormatter)

ax.tick_params(which = 'major', length = 8)         # 设置主刻度线长度
ax.tick_params(which = 'minor', length = 4)         # 设置副刻度线长度

ax.xaxis.grid(True, which = 'major', linestyle = (0, (8, 4)))            # 设置x方向网格线 (每个线段长8，间隔4，从0/8开始绘制)
ax.yaxis.grid(True, which = 'major', linestyle = (0, (8, 4)))            # 设置y方向网格线

# 坐标轴文字
ax.set_xlabel("X axis label", fontsize = 12)        # x坐标轴文字
ax.set_ylabel("Y axis label", fontsize = 12)        # y坐标轴文字

# 添加图像
x = np.arange(0.5, 3.3, 0.1)
y1 = -0.04658 * x ** 3 + 0.5494 * x ** 2 - 1.3151 * x + 1.415
y2 = 0.1545 * x ** 3 - 0.799 * x ** 2 + 0.4316 * x + 3.744
ax.plot(x, y1, color = 'r', alpha = 0.7, linewidth = 2.3, label = "Red Signal")
ax.plot(x, y2, color = 'b', alpha = 0.7, linewidth = 2.3, label = "Blue Signal")
ax.scatter(A, B, marker = 'o', color = 'w', edgecolors = 'k')

# 添加图例
ax.legend()

# 添加文字
ax.text(2.9, 2.7, "Grid", fontsize = 11, color = 'b')
ax.text(3.45, 3.45, "Legend", fontsize = 11, color = 'b')
ax.text(1.9, 3.8, "Title", fontsize = 11, color = 'b')
ax.text(0.05, 3.0, "Major tick", fontsize = 8, color = 'b')
ax.text(0.05, 3.75, "Minor tick", fontsize = 8, color = 'b')
ax.text(-0.45, 1.5, "Y axis label", fontsize = 8, color = 'b')
ax.text(1.8, -0.55, "X axis label", fontsize = 8, color = 'b')
ax.text(1.8, 0.35, "Line Plot", fontsize = 11, color = 'b')

# 添加箭头
ax.arrow(3.3, 0.41, -0.15, -0.33, ec = 'b', head_width = 0.04)
ax.arrow(3.6, 0.41, 0.33, -0.15, ec = 'b', head_width = 0.04)
ax.text(3.25, 0.5, "Splines", fontsize = 11, color = 'b')

# 保存
plt.savefig(fname = "Anacomy.pdf", format = "pdf")

# 显示
plt.show()
```
