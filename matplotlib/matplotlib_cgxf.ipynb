## 多个图形显示在一张图上

$ \text{在一张图上画出}y = sinx,\quad y = cosx^2, \quad x \in [0, 2\pi] $

```python
x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x**2)
plt.plot(x, y1, label='$sinx$', linewidth=2)
plt.plot(x, y2, label='$cosx^2$', linestyle='--')
plt.xlabel("$x$")
plt.ylabel("$y$", rotation=0)
plt.legend()
plt.show()
```

## 多个图单独显示

将屏幕开成三个子窗口，上两个，下一个，分别画

$ y = sinx \qquad y = cosx \qquad y = sinx^2 \qquad x \in [0, 2\pi] $

```python
x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(pow(x, 2))

ax1 = plt.subplot(2, 2, 1)
ax1.plot(x, y1, linestyle='-', color='r', linewidth=2.0, label='$sinx$')
ax1.legend()
ax2 = plt.subplot(2, 2, 2)
ax2.plot(x, y2, linestyle='--', color='y', linewidth=2.0, label='$cosx$')
ax2.legend()
ax3 = plt.subplot(2, 1, 2)
ax3.plot(x, y3, linestyle='--', color='b', linewidth=2.0, label='$sinx^2$')
ax3.legend()
plt.suptitle("三角函数")
plt.show()
```

### 三维空间曲面

绘制三维曲面$ z = sin(\sqrt{x^2 + y^2}) $的表面图和三维网格图

```python
# np.meshgrid()函数 (生成网格点坐标矩阵)
# 即把两个数组的笛卡尔积内的元素的第一、二个坐标分别放入两个矩阵中
x = np.array([0, 1, 2])
y = np.array([3, 4, 5])

X, Y = np.meshgrid(x, y)
print(X, '\n')
print(Y)
```

笛卡尔积

$ \textbf{a} = [0, 1, 2] \qquad \textbf{b} = [3, 4, 5] $

$ \textbf{a} \times \textbf{b} = \{<0,3>,<0,4>,<0,5>,<1,3>,<1,4>,<1,5>,<2,3>,<2,4>,<2,5>\} $

```python
#绘图
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)    #生成网格点坐标矩阵
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure(figsize=(12,16))    #设置绘图区的大小
# 表面图
ax1 = plt.subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$z$')
# 网格图
ax2 = plt.subplot(122, projection='3d')
ax2.plot_wireframe(X, Y, Z)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_zlabel('$z$')
plt.show()
```

### 等高线图

以$ z = x^2 + y^2 $为例

```python
#取点
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
#组成网格图
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
#设置画板大小
plt.figure(figsize=(8, 5))
#填充颜色，f即filled, levels是颜色层数
cor = plt.contourf(X, Y, Z, levels=6, cmap='cool')
#画等高线
contr = plt.contour(X, Y, Z, levels=6, colors='k')
#在等高线上标明高度值
plt.clabel(CS=contr, fontsize=15, colors='k')
#删去坐标轴
plt.xticks(())
plt.yticks(())
plt.title('等高线')
#添加颜色条
plt.colorbar(mappable=cor)
#显示
plt.show()
```

```python
#绘制三维图与等高图对比
plt.figure(figsize=(9, 6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='cool')
#设置坐标轴
plt.xticks(np.linspace(-10, 10, 7))
plt.yticks(np.linspace(-10, 10, 7))

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$z = x^2 + y^2$')
plt.show()
```

### 向量图
绘出速度向量$ (u, v) = (ycosx, ysinx) $的向量场

```python
x = np.linspace(0, 15, 11)
y = np.linspace(0, 10, 12)
#生成网格点
X, Y = np.meshgrid(x, y)
#速度向量
u = Y * np.cos(X)
v = Y * np.sin(X)
#向量图
plt.quiver(X, Y, u, v, color='r')
plt.title('速度向量图')
plt.show()
```

#### 莫比乌斯带

参数方程为: 
$$
\begin{cases}
    x = (2 + \frac{s}{2}cos\frac{t}{2})cost \\ 
    y = (2 + \frac{s}{2}cos\frac{t}{2})sint \\
    z = \frac{s}{2}sin\frac{t}{2}
\end{cases}
\qquad\quad t \in [0, 2\pi], \quad s \in [-1, 1]
$$

```python
#取点
t = np.linspace(0, 2*np.pi, 500)
s = np.linspace(-1, 1, 500)
#网格点
T, S = np.meshgrid(t, s)

x = (2 + S/2 * np.cos(T/2)) * np.cos(T)
y = (2 + S/2 * np.cos(T/2)) * np.sin(T)
z = S/2 * np.sin(T/2)

plt.figure(figsize=(10, 6))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)
#拉长z坐标轴
ax.set_zticks(np.linspace(-2, 2, 9))
ax.set_title('mobius')
plt.show()
```
