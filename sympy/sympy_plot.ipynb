# SymPy Plot

### plot(*args, show=True, **kwargs)

#### 绘制具有单个范围的单个表达式

plot(expr, range, **kwargs)

#### 绘制具有默认范围(-10, 10)的单个表达式

plot(expr, **kwargs)

#### 绘制具有单个范围的多个表达式

plot(expr1, expr2, ..., range, **kwargs)

#### 绘制具有多个范围的多个表达式

plot((expr1, range1), (expr2, range2), ..., **kwargs)

### kwargs

show: 默认Ture，当设置为False时函数图像不会显示

xlabel, ylabel: x，y坐标标签

line_color: 设置图像颜色

title: 设置标题

label: 设置图像标签，默认是表达式名字

xscale, yscale: 设置x，y轴缩放

xlim, ylim: 表示x，y坐标范围(min, max)

axis_center: 表示中心的坐标

annotations: 一个字典列表，指定所需注释的类型

markers: A list of dictionaries specifying the type the markers required

rectangles: 一个字典列表，指定要绘制的矩形的尺寸

fill: 一个字典列表，指定需要填充的颜色

**adaptive**: The default value is set to True. 

Set adaptive to False and specify nb_of_points if uniform sampling is required.

**depth**: Recursion depth of the adaptive algorithm

**nb_of_points**: Used when the adaptive is set to False. 

The function is uniformly sampled at nb_of_points number of points.

If the adaptive flag is set to True, this will be ignored

size: 指定完整图像大小

```python
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')
p1 = plot(1/2*x**2, (x, -5, 5), show=False, line_color='r')
p2 = plot(x, -x, (x, -5, 5), show=False)
# 使用extend()将第二张图添加到第一张图上
p1.extend(p2)
# 显示
p1.show()
```

```python
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')
p = plot(x, x**2, x**3, (x, -5.0, 5.0), show=False)
color = ['r', 'b', 'g']
labels = ['$x$', '$x^2$', '$x^3$']
for i in range(3):
    p[i].line_color = color[i]
    p[i].label = labels[i]
p.title = "Function"
p.ylabel = 'y'
p.legend = True
p.show()
```

### plot_parametric(*args, show=True, **kwargs)

参数方程绘图

**plot_parametric((expr_x, expr_y), range)**

**plot_parametric((expr_x, expr_y), ..., range)**

**plot_parametric((expr_x, expr_y, range), ...)**

关键词参数与上述plot函数大致相同

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot_parametric

u = symbols('u')
p = plot_parametric((cos(u), sin(u)), (u, -5, 5), show=False)
p.show()
```

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot_parametric

p = plot_parametric((cos(u), sin(u)), (u, cos(u)), 
                    (u, -10, 10), show=False)
p[0].line_color = 'g'
p[1].line_color = 'r'
p.show()
```

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot_parametric

p = plot_parametric((cos(u), sin(u), (u, -5, 5)), 
                    (cos(u), u, (u, -5, 5)), show=False)
p.title = 'Function'
p.show()
```

**绘图使用自适应算法递归抽样，以准确地绘制曲线。自适应算法使用两个点中点附近的随机点，必须进一步采样。**

**因此，由于是随机抽样，重复使用相同的plot命令可能会得到稍微不同的结果。**

```python
from sympy import symbols, cos, sin, pi
from sympy.plotting import plot_parametric

expr1 = (u, cos(2*pi*u)/2+1/2)
expr2 = (u, sin(2*pi*u)/2+1/2)
p = plot_parametric(expr1, expr2, (u, 0, 1), line_color='blue', show=False)
# 设置第二条曲线color
p[1].line_color = 'red'
# 显示图像
p.show()
```

### plot3d(*args, show=True, **kwargs) 绘制3d曲面图

**单个图像**

plot3d(expr, range_x, range_y, **kwargs)

**多个图像同一范围**

plot3d(expr1, expr2, range_x, range_y, **kwargs)

**多个图像不同范围**

plot3d((expr1, range_x, range_y), (expr2, range_x, range_y), ..., **kwargs)

### Keyword Arguments

**nb_of_points_x, y**: x, y范围在点的nb_of_points_x, y处均匀采样

**surface_color**: 指定图像表面颜色

**title**: plot title

**and so on**

```python
from sympy import symbols, sin, sqrt
from sympy.plotting import plot3d

x, y = symbols('x, y')
p = plot3d(sin(sqrt(x**2+y**2)), (x, -10, 10), (y, -10, 10), show=False)
p.show()
```

```python
# Multiple plots with different ranges(多图多范围)
from sympy import symbols
from sympy.plotting import plot3d

x, y = symbols('x y')
p = plot3d((x**2+y**2, (x, -5, 5), (y, -5, 5)), 
        (x+y, (x, -3, 3), (y, -3, 3)), show=False)
p.show()
```

### plot3d_parametric_line(*args, show=True, **kwargs) 3d曲线

#### Single Plot 

** plot3d_parametric_line(expr_x, expr_y, expr_z, range, \*\*kwargs)**

#### Multiple Plots

** plot3d_parametric_line((expr_x, expr_y, expr_z, range), ..., \*\*kwargs) **

```python
# 单条曲线
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_line

u = symbols('u')
p = plot3d_parametric_line(cos(u), sin(u), u, (u, -5, 5), show=False, line_color='r')
p.show()
```

```python
# 多条曲线
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_line

u = symbols('u')
p = plot3d_parametric_line((cos(u), sin(u), u, (u, -5, 5)),
                            (sin(u), u**2, u, (u, -5, 5)), show=False)
# Set Curve Color
p[0].line_color = 'r'
p[1].line_color = 'b'
p.show()
```

### plot3d_parametric_surface(*args, show=True, **kwargs) 3d曲面参数方程

#### Single plot

**plot3d_parametric_surface(expr_x, expr_y, expr_z, range_u, range_v, \*\*kwargs)**

#### Multiple plots

**plot3d_parametric_surface((expr_x, expr_y, expr_z, range_u, range_v), ..., \*\*kwargs)**

```python
from sympy import symbols, cos, sin, pi
from sympy.plotting import plot3d_parametric_surface
u, v = symbols('u v')

# 半径为2的球
p = plot3d_parametric_surface(2*sin(u)*cos(v), 2*sin(u)*sin(v), 2*cos(u), (u, 0, pi), (v, 0, 2*pi), show=False)
p.show()
```

### plot_implicit(expr, x_var=None, y_var=None, adaptive=True, depth=0, points=300, line_color='blue', show=True, **kwargs)

** 隐函或不等式数绘图 **



expr: 需要画的方程或不等式

x_var, y_var: 给定符号和范围为(symbol, min, max)

adaptive: The default value is set to True. It has to be set to False if you want to use a mesh grid

depth: The depth of recursion for adaptive mesh grid\默认为0，范围是(0, 4)

points: The number of points if adaptive mesh grid is not used. Default value is 300.

title, xlabel, ylabel, line_color 不再赘述

```python
from sympy import symbols, Eq
from sympy.plotting import plot_implicit

x, y = symbols('x, y')
p1 = plot_implicit(Eq(x**2 + y**2, 5), line_color='r')
```

```python
# 设置x, y范围
p2 = plot_implicit(Eq(x**2 + y**2, 5), (x, -3, 3), (y, -3, 3), line_color='r')    
```

```python
# 设置递归参数深度(变淡了，滑稽)
p3 = plot_implicit(Eq(x**2 + y**2, 5), line_color='r', depth=2) 
```

```python
# 不使用adaptive
p4 = plot_implicit(Eq(x**2 + y**2, 5), (x, -4, 4), (y, -3, 3), line_color='r', adaptive=False) 
```
